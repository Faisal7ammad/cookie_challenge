```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cookies_fixed = pd.read_csv("../data/cookies_fixed.csv",index_col = 0)
cookies_val_fixed = pd.read_csv("../data/cookies_validate_fixed.csv",index_col = 0)
cookies_fixed
```

```python
from sklearn.model_selection import cross_validate as cv
```

```python
X = cookies_fixed.drop(columns = ['quality', 'raisins','peanut_butter', 'oats', "chocolate"])
y = cookies_fixed['quality']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.tree import DecisionTreeRegressor


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def performance_model(y_train, y_pred_train):

    # Model validation y_train - y_pred_train
    ME_train = np.mean(y_train-y_pred_train)

    MAE_train = mean_absolute_error(y_train,y_pred_train)
    
    MSE_train = mean_squared_error(y_train,y_pred_train)

    RMSE_train = np.sqrt(MSE_train)

    MAPE_train = np.mean((np.abs(y_train-y_pred_train) / y_train)* 100.)

    R2_train = r2_score(y_train,y_pred_train)

    performance = pd.DataFrame({'Error_metric': ['Mean error','Mean absolute error','Mean squared error',
                                             'Root mean squared error','Mean absolute percentual error',
                                             'R2'],
                            'Train': [ME_train, MAE_train, MSE_train, RMSE_train, MAPE_train, R2_train]})
    pd.options.display.float_format = '{:.2f}'.format


    df_train = pd.DataFrame({'Real_value': y_train, 'Predicted_value': y_pred_train})

    return performance, df_train

#performance, _ = performance_model(y, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(RandomForestRegressor(), {'n_estimators':(list(range(175,401,25))), "max_depth":list(range(10,21))},cv = 5)
clf.fit(X,y)
```

```python
clf.best_score_
```

```python
clf.best_estimator_
```

```python
clf.best_params_
```

```python
x_test = cookies_val_fixed.drop(columns = ["quality", 'raisins','peanut_butter', 'oats', "chocolate"])
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
cookies_val_fixed['quality'] = list(y_pred)
```

```python

# evaluate gradient boosting ensemble for regression
from numpy import mean
from numpy import std
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor()
# define the evaluation procedure
cv = RepeatedKFold(n_splits=5, n_repeats=10, random_state=1)
# evaluate the model
n_scores = cross_val_score(model, X, y, scoring='r2', cv=cv, n_jobs=-1)
# report performance
print('r2: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
CLF = GridSearchCV(GradientBoostingRegressor(), {"max_depth":list(range(3,6)), 'n_estimators':(list(range(250,301,50)))},cv = 5)
CLF.fit(X,y)
```
